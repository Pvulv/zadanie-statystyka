```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
mu = 3
sigma = 1

n_100 = 100
n_10000 = 10000

data_100 = np.random.normal(mu, sigma, n_100)
data_10000 = np.random.normal(mu, sigma, n_10000)
```

### Średnia

```python
mean_100 = np.round(np.mean(data_100))
mean_100
```

```python
mean_10000 = np.round(np.mean(data_10000))
mean_10000
```

### Mediana

```python
median_100 = np.median(data_100)
median_100
```

```python
median_10000 = np.median(data_10000)
median_10000
```

### Moda

```python
mode_100 = stats.mode(data_100)
mode_100
```

```python
mode_10000 = stats.mode(data_10000)
mode_10000
```

Używając losowych wartości można się spodziewać, że moda nie będzie występować. (count = 1)

### Kwartyle data_100

```python
q0_100 = np.quantile(data_100, 0.0)
q1_100 = np.round(np.quantile(data_100, 0.25),2)
q3_100 = np.round(np.quantile(data_100, 0.75),2)
q4_100 = np.round(np.quantile(data_100, 1.0),2)
print(f'Q0: {q0_100}')
print(f'Q1: {q1_100}')
print(f'Q2: {median_100}')
print(f'Q3: {q3_100}')
print(f'Q4: {q4_100}')
```

```python
plt.figure(figsize=(15,10))
plt.hist(data_100, bins=20)
plt.axvline(x = mean_100, color='red', label = 'Średnia', linestyle = '--')
plt.axvline(x = median_100, color='purple', label = 'Mediana', linestyle = '--')
plt.axvline(x = mode_100[0], color='orange', label = 'Moda', linestyle = '--')
plt.axvline(x = q0_100, color='black', label = 'Min', alpha = 0.5)
plt.axvline(x = q1_100, color='black', label = 'Q1', alpha = 0.5)
plt.axvline(x = q3_100, color='black', label = 'Q3', alpha = 0.5)
plt.axvline(x = q4_100, color='black', label = 'Max', alpha = 0.5)
plt.legend(loc='upper right')
plt.show()
```

### Kwartyle data_10000

```python
q0_10000 = np.quantile(data_10000, 0.0)
q1_10000 = np.round(np.quantile(data_10000, 0.25),2)
q3_10000 = np.round(np.quantile(data_10000, 0.75),2)
q4_10000 = np.round(np.quantile(data_10000, 1.0),2)
print(f'Q0: {q0_10000}')
print(f'Q1: {q1_10000}')
print(f'Q2: {median_10000}')
print(f'Q3: {q3_10000}')
print(f'Q4: {q4_10000}')
```

```python
plt.figure(figsize=(15,10))
plt.hist(data_10000, bins=20)
plt.axvline(x = mean_10000, color='red', label = 'Średnia', linestyle = '-.')
plt.axvline(x = median_10000, color='purple', label = 'Mediana', linestyle = '--')
plt.axvline(x = mode_10000[0], color='orange', label = 'Moda', linestyle = '--')
plt.axvline(x = q0_10000, color='black', label = 'Min', alpha = 0.5)
plt.axvline(x = q1_10000, color='black', label = 'Q1', alpha = 0.5)
plt.axvline(x = q3_10000, color='black', label = 'Q3', alpha = 0.5)
plt.axvline(x = q4_10000, color='black', label = 'Max', alpha = 0.5)
plt.legend(loc='upper right')
plt.show()
```

### Standaryzacja

```python
scaler_standardized = StandardScaler()
standardized_100 = scaler_standardized.fit_transform(data_100.reshape(-1, 1))
plt.hist(standardized_100, bins=20)
plt.axvline(x = standardized_100.mean(), color='red', label = 'Średnia', linestyle = '--')
plt.show()
print(f'Średnia: {standardized_100.mean()}')
print(f'Odchylenie standardowe: {np.std(standardized_100, ddof=1)}')
```

```python
scaler_standardized = StandardScaler()
standardized_10000 = scaler_standardized.fit_transform(data_10000.reshape(-1, 1))
plt.hist(standardized_10000, bins=20)
plt.axvline(x = standardized_10000.mean(), color='red', label = 'Średnia', linestyle = '--')
plt.show()
print(f'Średnia: {standardized_10000.mean()}')
print(f'Odchylenie standardowe: {np.std(standardized_10000, ddof=1)}')
```

```python
plt.figure(figsize=(12, 8))
plt.hist(data_100, bins=20, alpha=0.7, label='100 danych', color='blue', density=True)
plt.hist(data_10000, bins=50, alpha=0.7, label='10 000 danych', color='orange', density=True)
plt.title('Rozkład normalny: średnia = 3, odch. std = 1')
plt.xlabel('Wartość')
plt.legend()
plt.show()
```

Wnioski : Zbiór posiadający 10000 wartości jest dokładniejszy i więcej można z niego odczytać. W zależności co chcemy uzyskać 10000 wartości będzie mieć mniejszy rachunek błędu. Nie zależnie od losowości zbiorów zbiór 100 wartości będzie zawsze "ukrucony" z większym błędem nawet przy takim samym odchyleniu standardowym oraz średnej.
